```python
import requests
import base64

def find_token():
    clientId = "18e2c5fe357d4a67bb56e304680da8aa"
    clientSecret = "1f4dcd21d5eb48b9a803caa628441520"
    auth_url = "https://accounts.spotify.com/api/token"

    # Codifica o clientId e clientSecret em base64
    auth_str = f"{clientId}:{clientSecret}"
    b64_auth_str = base64.b64encode(auth_str.encode()).decode()

    # Define os headers e os dados da requisição
    headers = {
        "Authorization": f"Basic {b64_auth_str}"
    }
    data = {
        "grant_type": "client_credentials"
    }

    # Fazendo a requisição POST para obter o token
    response = requests.post(auth_url, headers=headers, data=data)

    # Verificando se a requisição foi bem-sucedida
    if response.status_code == 200:
        token_info = response.json()
        return token_info['access_token']
    else:
        raise Exception(f"Erro ao obter o token: {response.status_code} - {response.json()}")

def search_tracks_by_genre(genre, limit, market="BR"):
    token = find_token()

    # Construindo a URL da requisição
    url = f"https://api.spotify.com/v1/search?q=genre%3D{genre}&type=track&market={market}&limit={limit}"

    # Cabeçalho com a autorização
    headers = {
        "Authorization": f"Bearer {token}"
    }

    # Fazendo a requisição GET
    response = requests.get(url, headers=headers)

    # Verificando o status da resposta
    if response.status_code == 200:
        # Se a resposta for bem-sucedida, retorna os dados
        return response.json()
    else:
        # Se houver um erro, retorna o código de status e a mensagem de erro
        return {"error": response.status_code, "message": response.json()}

# Exemplo de uso
result = search_tracks_by_genre("rock", 2)
print(result)
```

```python
def get_track_info(track_id):
    # Obtendo o token usando a função find_token()
    token = find_token()

    # Construindo a URL da requisição
    url = f"https://api.spotify.com/v1/tracks/{track_id}"

    # Cabeçalho com a autorização
    headers = {
        "Authorization": f"Bearer {token}"
    }

    # Fazendo a requisição GET
    response = requests.get(url, headers=headers)

    # Verificando o status da resposta
    if response.status_code == 200:
        # Se a resposta for bem-sucedida, retorna os dados
        return response.json()
    else:
        # Se houver um erro, retorna o código de status e a mensagem de erro
        print(response)
        erro = {"error": response.status_code, "message": response.json()}
        return erro;
def get_track_audioFeatures(track_id):
    # Obtendo o token usando a função find_token()
    token = find_token()

    # Construindo a URL da requisição
    url = f"https://api.spotify.com/v1/audio-features/{track_id}"

    # Cabeçalho com a autorização
    headers = {
        "Authorization": f"Bearer {token}"
    }

    # Fazendo a requisição GET
    response = requests.get(url, headers=headers)

    # Verificando o status da resposta
    if response.status_code == 200:
        # Se a resposta for bem-sucedida, retorna os dados
        return response.json()
    else:
        # Se houver um erro, retorna o código de status e a mensagem de erro
        print(response)
        erro = {"error": response.status_code, "message": response.json()}
        return erro;
```

```python
import time

def createObjectByGenre(genre, limit):
    result = search_tracks_by_genre(genre, limit)
    objects_list = []
    print(genre)
    for item in result['tracks']['items']:
        track_id = item['id']
        time.sleep(0.3)
        track = get_track_info(track_id)
        features = get_track_audioFeatures(track_id)

        # Cria um objeto com as informações necessárias
        track_object = {
            "album_type": track['album']['album_type'],
            "release_date": track['album']['release_date'],
            "explicit": track['explicit'],
            "popularity": track['popularity'],
            "acousticness": features['acousticness'],
            "danceability": features['danceability'],
            "duration_ms": features['duration_ms'],
            "energy": features['energy'],
            "id": features['id'],
            "instrumentalness": features['instrumentalness'],
            "key": features['key'],
            "liveness": features['liveness'],
            "loudness": features['loudness'],
            "mode": features['mode'],
            "speechiness": features['speechiness'],
            "time_signature": features['time_signature'],
            "valence": features['valence'],
            "genre": genre  # Adiciona o gênero ao objeto
        }

        # Adiciona o objeto à lista
        objects_list.append(track_object)

    # Retorna a lista de objetos
    return objects_list
```

```python
genres = {"rock", "sertanejo", "funk", "pagode", "pop", "mpb", "forro", "hip-hop", "reggae", "k-pop", "gospel", "electronic"}
# Exemplo de uso
limit = 50

def createObjectsForGenres(genres, limit):
    objects_list = []
    for genre in genres:
        objects = createObjectByGenre(genre, limit)
        objects_list.extend(objects)
    return objects_list

objects = createObjectsForGenres(genres, limit)
print(objects)
```
